```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#cargar el dataset
df = pd.read_csv('C:/UNAD/wine_quality_red.csv')
```

```python
print(df.columns)
```

```python
#mostrar las primeras filas del dataset
print("Primeras filas del Dataset")
display(df.head(20))
```

```python
#Informacion general 
print("\nInformacion general:")
print(df.info())

#Estadisticas
print("\nEstadisticas descriptivas")
print(df.describe())

#Verificar la distribucion de la variable objetivo la cual es la calidad 
plt.figure(figsize=(8,5))
sns.countplot(x='quality', data=df, hue='quality', palette='Set2', legend=False)
plt.title('Distribucion de la Calidad del Vino')
plt.xlabel('Calidad')
plt.ylabel('Frecuencia')
plt.show()

#Matriz de Correlacion
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt="2f")
plt.title('Matriz de Correlacion entre variables')
plt.show()
```

```python
print ("2. Procesar los datos limpiándolos, tratando valores faltantes y transformandolos segun sea necesario.")
```

```python
#Verificar valores faltantes
print("Valores faltantes por columna:")
print(df.isnull().sum())
```

```python
print ("3. Seleccionar las caracteristicas más relevantes para entrenar el modelo utilizando seleccion de caracteristicas.")
```

```python
from sklearn.ensemble import ExtraTreesClassifier

x = df.drop('quality', axis=1)
y = df['quality']

model = ExtraTreesClassifier()
model.fit(x, y)

#Mostrar las caracteristicas más relevantes
feat_importances = pd.Series(model.feature_importances_, index=x.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.title("Caracteristicas más relevantes")
plt.show()
```

```python
print ("4. Dividir el dataset en Train y Test para evaluar correctamente el modelo.")
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y)

print(f"Train shape: {x_train.shape}, Test shape: {x_test.shape}")
```

```python
print ("5. Entrenar el modelo configurando los diferentes hiperparametros.")
```

```python
from sklearn.tree import DecisionTreeClassifier

#Crear y entrenar el modelo
dt_model = DecisionTreeClassifier(max_depth=5, random_state=42)
dt_model.fit(x_train, y_train)
```

```python
print ("6. Evaluar el desempeño del modelo en el conjunto Test con métricas como precisión, recall, F1-score, etc.")
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

y_pred = dt_model.predict(x_test)

print("Reporte de Clasificacion:")
print(classification_report(y_test, y_pred, zero_division=0))
```

```python
print ("7. Realizar las diferentes graficas que permitan visualizar los resultados del modelo.")
```

```python
#Matriz de confusion
print("Matriz de Confusion:")
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt="d", cmap='Blues')
plt.xlabel("Predicho")
plt.ylabel("Real")
plt.title("Matriz de Confusion")
plt.show()

print(f"Precision del modelo: {accuracy_score(y_test, y_pred):.2f}")
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20,10))
plot_tree(dt_model, feature_names=x.columns, class_names=[str(i) for i in sorted(y.unique())], filled=True)
plt.title("Visualizacion del Arbol de Decision")
plt.show()
```
